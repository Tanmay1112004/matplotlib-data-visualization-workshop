```python
import pandas as pd
data = {
    "Month" : ['January', 'February', 'March', 'April', 'May', 'June',],
    "Sales" : [10000, 12000, 15000, 13000, 17000, 16000],
    "Profit" : [2000, 3000, 4000, 2500, 3500, 3000]
}
df = pd.DataFrame(data)
print(df)
```

```python
df[['Month', 'Sales']]
```

# 1 - LINE PLOT MONTHLY SALES


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 3))
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='blue', label='Sales')
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.legend()
plt.grid(True)
plt.show()
```

#2 - BARPLOT MONTH VS PROFIT

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 3))
width = 0.3
plt.bar(df['Month'], df['Sales'], width=width, color='skyblue', label='Sales')
plt.bar(df['Month'], df['Profit'], width=width, color='green', label='Profit', bottom=df['Sales'])
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Sales and Profit Trends by Months')
plt.grid(True)
plt.legend()
# Get the current figure and apply tight_layout
fig = plt.gcf()  # Get current figure
fig.tight_layout()  # Apply tight layout
plt.show()
```

#3 - PIE CHART PROFIT VS MONTHS

```python
plt.figure(figsize=(10, 5))
plt.pie(df['Profit'], labels=df['Month'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightsalmon', 'lightseagreen', 'lightskyblue', 'lightsteelblue', 'lightyellow', 'lightcyan', 'lightgoldenrodyellow', 'lightpink'])
plt.title('Profitby Months')
plt.show()
```


#4 - SCATTER PLOT

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Sales'], df['Profit'], color='green', s = 100, edgecolors='black')
plt.title('Sales vs Profit')
plt.xlabel('Sales ')
plt.ylabel('Profit ')
plt.tight_layout()
plt.grid(True)
plt.show()
```

#5 - HISTOGRAM

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Sales'], bins=5, color='purple', edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#6 - BOX PLOT

```python
plt.figure(figsize=(8, 4))
plt.boxplot(df['Profit'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Sales')
plt.xlabel('Sales ($)')
plt.show()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Month" : ['January', 'February', 'March', 'April', 'May', 'June',],
    "Sales" : [10000, 12000, 15000, 13000, 17000, 16000],
    "Profit" : [2000, 3000, 4000, 2500, 3500, 3000]
}
df = pd.DataFrame(data)

def generate_plot(plot_type):
    fig = plt.figure(figsize=(8, 5))

    if plot_type == 'Line Plot':
      plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='blue', label='Sales')
      plt.title('Sales Trends Over Months')
      plt.xlabel('Month')
      plt.ylabel('Sales ($)')
      plt.legend()
      plt.grid(True)

    elif plot_type == 'stacked Bar Chart':
      fig.set_size_inches(10, 6)
      width = 0.3
      plt.bar(df['Month'], df['Sales'], width=width, color='skyblue', label='Sales')
      plt.bar(df['Month'], df['Profit'], width=width, color='green', label='Profit', bottom=df['Sales'])
      plt.xlabel('Month')
      plt.ylabel('Amount ($)')
      plt.title('Sales and Profit Trends by Months')
      plt.grid(True)
      plt.legend()
      fig = plt.gcf()  # Get current figure
      fig.tight_layout()  # Apply tight layout


    elif plot_type == 'Pie Chart':
        fig.set_size_inches(7,7)
        plt.pie(df['Profit'], labels=df['Month'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
        plt.title('Profit by Months')


    elif plot_type == 'Scatter Plot':
      plt.scatter(df['Sales'], df['Profit'], color='green', s=100, edgecolors='black')
      plt.title('Sales vs Profit')
      plt.xlabel('Sales ($)')
      plt.ylabel('Profit ($)')
      plt.grid(True)

    elif plot_type == 'Histogram':
      plt.hist(df['Sales'], bins=5, color='purple', edgecolor='black')
      plt.title('Distribution of Sales')
      plt.xlabel('Sales ($)')
      plt.ylabel('Frequency')

    elif plot_type == 'Box Plot':
      plt.boxplot(df['Profit'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
      plt.title('Box Plot of Profit')
      plt.xlabel('Profit ($)')

    # Correct the indentation of this line
    plt.tight_layout()
    return fig

# gradio ui

demo = gr.Interface(
    fn=generate_plot,
    inputs=gr.Dropdown(['Line Plot', 'stacked Bar Chart', 'Pie Chart', 'Scatter Plot', 'Histogram', 'Box Plot'], label='Select Plot Type'),
    outputs=gr.Plot(label = 'Sales data Visualization'),
    title='Sales & Profit visual insight',
    description='Select a plot type to visualize sales data.'
)

demo.launch()
```
